```python
import openpyxl
import pandas as pd
from tabulate import tabulate


df = openpyxl.load_workbook("C:/Users/ACER NITRO/Downloads/datos.xlsx")


hoja = df.active


data = []

headers = ["Row Number"] + ["Customer ID", "Gender", "Age", "Under 30", "Senior Citizen", "Married", "Dependents", 
           "Number of Dependents", "Country", "State", "City", "Zip Code", "Latitude", "Longitude", 
           "Population", "Quarter", "Referred a Friend", "Number of Referrals", "Tenure in Months", 
           "Offer", "Phone Service", "Avg Monthly Long Distance Charges", "Multiple Lines", 
           "Internet Service", "Internet Type", "Avg Monthly GB Download", "Online Security", 
           "Online Backup", "Device Protection Plan", "Premium Tech Support", "Streaming TV", 
           "Streaming Movies", "Streaming Music", "Unlimited Data", "Contract", "Paperless Billing", 
           "Payment Method", "Monthly Charge", "Total Charges", "Total Refunds", 
           "Total Extra Data Charges", "Total Long Distance Charges", "Total Revenue", 
           "Satisfaction Score", "Customer Status", "Churn Label", "Churn Score", "CLTV", 
           "Churn Category", "Churn Reason"]


for index, row in enumerate(hoja.iter_rows(min_row=2, max_row=hoja.max_row, values_only=True), start=1):

    if any(row):
      
        data.append([index] + list(row))


df_data = pd.DataFrame(data, columns=headers)


print(df_data.shape)


datos = tabulate(df_data, headers=headers, tablefmt='fancy_grid')


print(datos)





```

```python
df_data.dropna(inplace=True)
df_data.info()
```

```python
cols_cat = ["Customer ID", "Gender", "Married", "Dependents"
        , "City", "Referred a Friend", "Offer", "Phone Service", "Multiple Lines", 
            "Internet Service", "Internet Type", "Online Security", "Online Backup", "Device Protection Plan", 
            "Premium Tech Support", "Streaming TV", "Streaming Movies", "Streaming Music", "Unlimited Data", 
            "Contract", "Paperless Billing", "Payment Method", "Total Long Distance Charges", "Total Revenue" 
             , "Churn Category", "Churn Reason"]


cols_to_keep = []


for col in cols_cat:
    num_unique = df_data[col].nunique()  
    print(f'Columna {col}: {num_unique} subniveles')
    
    
    if num_unique > 1:
        cols_to_keep.append(col)


df_filtered = df_data[cols_to_keep]


print(f"Columnas conservadas: {cols_to_keep}")
print(df_filtered)
df_filtered.shape
```

```python
print(f'Tamaño del set antes de eliminar las filas repetidas: {df_data.shape}')
df_data.drop_duplicates(inplace=True)
print(f'Tamaño del set después de eliminar las filas repetidas: {df_data.shape}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols_num = ["Row Number", "Age", "Number of Dependents", "Zip Code", 
            "Latitude", "Longitude", "Population", "Number of Referrals",
            "Tenure in Months", "Avg Monthly Long Distance Charges", "Avg Monthly GB Download", 
            "Monthly Charge", "Total Charges", "Total Refunds", "Total Extra Data Charges", 
            "Satisfaction Score", "Churn Score", "CLTV"]


n_cols = 5  #
n_rows = (len(cols_num) + n_cols - 1) // n_cols  


fig, ax = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20, n_rows * 5))
fig.subplots_adjust(hspace=0.5, wspace=0.3)  


for i, col in enumerate(cols_num):
    row = i // n_cols
    col_idx = i % n_cols
    sns.boxplot(x=df_data[col], ax=ax[row, col_idx])
    ax[row, col_idx].set_title(col)
    ax[row, col_idx].set_xticklabels(ax[row, col_idx].get_xticklabels(), rotation=45, ha='right', fontsize=10)  

for j in range(len(cols_num), n_rows * n_cols):
    fig.delaxes(ax.flatten()[j])


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


top_cities = df_data['City'].value_counts().nlargest(20).index  


df_top_cities = df_data[df_data['City'].isin(top_cities)]


fig, ax = plt.subplots(figsize=(15, 8))


sns.countplot(x='City', data=df_top_cities, ax=ax)
ax.set_title('Top 20 Cities')


ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')


plt.show()
```
